```python
def generateC(size) -> int:
    y = np.random.rand(size)
    # assuming cdf is x^2, inverse should be sqrt(x)
    return np.power(y, 1/2)


bid1 = 0.5216 # we get scuba gear and pay 0.7 to the buyer
bid2 = 0.7824 # we get scuba gear and pay 0.9 to the buyer

# Note: Say the buyer has a c of 0.6, then he will only buy if the bid is higher than his value of c

maxits = 10000
total = 0
for i in range(maxits):
    c = generateC(1)
    if bid1 > c:
        total = total  + 1-bid1
    elif bid2 > c:
        total = total + 1-bid2

print("avg profit", total/maxits)

# test = np.random.standard_normal(500)
# plt.hist(generateC(5000000), 100)
```
