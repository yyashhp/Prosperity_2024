```python
data_start_2.index = range(10001, 20002)
data_start_3.index = range(20002, 30003)

data_start = pd.concat([data_start_1, data_start_2, data_start_3], axis = 0)
data_start.index.values


```

```python
cols = str(data_start.columns.values).lstrip('[').rstrip(']').split(';')
cols[0] = 'timestamp'
cols[-1] = 'DAY'
cols
```

```python
#times = []
orchs = []
trans = []
exp = []
imp = []
sun = []
humid = []
#day = []
```

```python
def get_data(row):
    datas = str(row['timestamp;ORCHIDS;TRANSPORT_FEES;EXPORT_TARIFF;IMPORT_TARIFF;SUNLIGHT;HUMIDITY;DAY']).split(';')
    #times.append(datas[0])
    orchs.append(float(datas[1]))
    trans.append(float(datas[2]))
    exp.append(float(datas[3]))
    imp.append(float(datas[4]))
    sun.append(float(datas[5]))
    humid.append(float(datas[6]))
data_start.apply(get_data, axis = 1)
    
```

```python
times = range(0,3000300,100)
len(data_start)
```

```python
import pandas as pd

# Define the data
data = {
    'timestamp': times,
    'ORCHIDS': orchs,
    'TRANSPORT_FEES': trans,
    'EXPORT_TARIFF': exp,
    'IMPORT_TARIFF': imp,
    'SUNLIGHT': sun,
    'HUMIDITY': humid,
}

# Create the DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
df
```

```python
df.index = times
df
```
