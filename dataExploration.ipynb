```python
import numpy as np
import pandas as pd

# Assuming your data is stored in a DataFrame called 'df'
# and the target variable (next day's price) is 'mid_price'

# Create dummy variables for each timestamp
df_with_dummies = pd.get_dummies(df, columns=['timestamp'])

# Selecting predictor variables including dummy variables
predictor_cols = ['product','bid_price_1', 'bid_volume_1', 'bid_price_2', 'bid_volume_2', 
                  'bid_price_3', 'bid_volume_3', 'ask_price_1', 'ask_volume_1', 'ask_price_2', 
                  'ask_volume_2', 'ask_price_3', 'ask_volume_3']

# Remove original timestamp predictor column
predictor_cols = [col for col in predictor_cols if col != 'timestamp']

X = df_with_dummies[predictor_cols + list(df_with_dummies.columns[df_with_dummies.columns.str.startswith('timestamp_')])].values[1:]
y = df_with_dummies['mid_price'].diff().shift(-1).dropna().values

# Add a column of ones for the intercept term
X = np.column_stack([np.ones(len(X)), X])

# Calculate the coefficients using linear algebra
coefficients = np.linalg.lstsq(X, y, rcond=None)[0]

# Printing the coefficients
print("Intercept:", coefficients[0])
print("Coefficients for predictor variables:", coefficients[1:])
```

```python
print(np.linalg.lstsq(X, y, rcond=None)).
```

```python
import pickle

# Save coefficients using pickle
with open('linear_regression_model.pkl', 'wb') as f:
    pickle.dump(coefficients, f)
```
