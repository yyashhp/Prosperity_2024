```python
import statistics
from statistics import mean
```

```python
def get_stats(prices):
    print(f" \t       Mean: {mean(prices)} \n  \
             Min: {min(prices)} \n  \
             Max: {max(prices)} \n  \
             median: {statistics.median(prices)} \n  \
             std: {statistics.stdev(prices)} \n  \
             length: {len((prices))} \n")
```

```python
get_stats(amethyst_df['mid_price'])
```

```python
get_stats(starfruit_df['mid_price'])
```
