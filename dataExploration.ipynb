```python
import statistics
from statistics import mean
```

```python
def get_stats(prices):
    print(f" \t       Mean: {mean(prices)} \n  \
             min: {min(prices)} \n  \
             max: {max(prices)} \n  \
             median: {statistics.median(prices)} \n  \
             std: {statistics.stdev(prices)} \n  \
             length: {len((prices))} \n")
```

```python
get_stats(adf['mid_price'])
```
